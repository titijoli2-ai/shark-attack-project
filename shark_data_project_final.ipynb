# Shark Attack Data - Cleaning & Analysis

## Business Case
**The Problem:** Since the release of Jaws in 1975, sharks have been widely feared and misunderstood. In reality, only a small fraction of shark species pose any risk to humans, while sharks play a vital role in maintaining healthy ocean ecosystems.

**Our Goal:** Analysing the Global incident log of shark attacks and show that the likelihood of a fatality increases only when the shark is provoked.

**Key Questions We Answer:**
1. What proportion of shark incidents are unprovoked vs. provoked?
2. What are the types of the incidents?
3. How many incidents are actually fatal?
4. Which species are involved — and are they truly dangerous?
5. Which species appear in fatal vs. non-fatal cases?
6. Has the number of incidents increased over time — and if so, why?
7. Does gender play a role in shark incident patterns?
8. What are the types of the incidents?

**Scope:** USA only (largest dataset with most reliable records)

**The Story:** Sharks are not the mindless killers the media portrays. The data shows that most encounters are non-fatal, some are provoked by humans, and the species most often involved rarely cause serious harm. The rise in incidents over time reflects human enterance into shark habitat is not increasing shark aggression.

```python
# Let's clean the fatal column

# Step 1: Check how many unique items there are in this column
shark_clean_usa_filtered_df["fatal y/n"].unique()

# Step 2: Count how many values in the column. Include dropna=False so it shows missing values too.
shark_clean_usa_filtered_df["fatal y/n"].value_counts(dropna=False)

# Step 3: Replace UNKNOWN with NaN
shark_clean_usa_filtered_df["fatal y/n"] = shark_clean_usa_filtered_df["fatal y/n"].replace("UNKNOWN", np.nan)

# Step 4: Make all values a string (to deal with integers), remove white space and make them all uppercase (to deal with y & n) 
shark_clean_usa_filtered_df["fatal y/n"] = (
    shark_clean_usa_filtered_df["fatal y/n"]
    .astype("string")
    .str.strip()
    .str.lower()
)

# Step 5: Made a decision to only keep Y or N and add everything else to NaN, using a loop
valid_values = ["y", "n"]

# NB: .items() returns 2 values (is a tuple), the index and the Y/N (row, column) so you need to include BOTH in the for loop
for index, value in shark_clean_usa_filtered_df["fatal y/n"].items():

    # Added pd.notna(value) because when I converted to a string, pandas created NA values and was giving a "boolean value of NA is ambiguous" error
    if pd.notna(value) and value not in valid_values:

        # Using .loc to select a row & column. So for each row set the column value to NaN
        shark_clean_usa_filtered_df.loc[index,"fatal y/n"] = np.nan
        

shark_clean_usa_filtered_df["fatal y/n"].value_counts(dropna=False)
```

```python
# Step 1 - Cleaning by using strip() (whitespace) in Species column
# This prevents mismatches caused by invisible spaces
shark_clean_usa_filtered_df['species'] = shark_clean_usa_filtered_df['species'].str.strip()
```

```python
# Step 2 - Convert all Species values to lowercase to standardizes the text.
shark_clean_usa_filtered_df['species'] = shark_clean_usa_filtered_df['species'].str.lower()
```

```python
# Step 3 - Fill missing species values with 'unknown'
# We keep these rows in the dataset since the incident data is still valid
# They will be grouped into 'Other / Unknown' in later stage
shark_clean_usa_filtered_df['species'] = shark_clean_usa_filtered_df['species'].fillna('unknown')
```

```python
# Step 4 - Classify species into 4 groups: White Shark, Tiger Shark, Bull Shark, Other/Unknown
# These 3 species are the most commonly cited in shark incident reports.
# All other species and unknowns are grouped into 'Other / Unknown' to simplify the analysis.
# We use str.contains() to catch variations like 'white shark, 12ft' or 'suspected white shark'
shark_clean_usa_filtered_df.loc[shark_clean_usa_filtered_df['species'].str.contains('white'), 'species'] = 'White Shark'
shark_clean_usa_filtered_df.loc[shark_clean_usa_filtered_df['species'].str.contains('tiger'), 'species'] = 'Tiger Shark'
shark_clean_usa_filtered_df.loc[shark_clean_usa_filtered_df['species'].str.contains('bull'), 'species'] = 'Bull Shark'
```

```python
# Step 5 - Group everything that is not White, Tiger or Bull into 'Other / Unknown'
not_white = shark_clean_usa_filtered_df['species'] != 'White Shark'
not_tiger = shark_clean_usa_filtered_df['species'] != 'Tiger Shark'
not_bull = shark_clean_usa_filtered_df['species'] != 'Bull Shark'

condition_species = not_white & not_tiger & not_bull
shark_clean_usa_filtered_df.loc[condition_species, 'species'] = 'Other / Unknown'
```

```python
shark_clean_usa_filtered_df['species'].value_counts()
```

```python
# Fatality Rate - the most powerful information for our narrative
# This shows what percentage of all shark incidents in the USA resulted in a fatality

total_incidents = len(shark_clean_usa_filtered_df)
fatal_incidents = (shark_clean_usa_filtered_df["fatal y/n"] == "y").sum()
fatality_rate = (fatal_incidents / total_incidents * 100).round(2)

print(f"Total shark incidents in USA: {total_incidents}")
print(f"Fatal incidents: {fatal_incidents}")
print(f"Fatality rate: {fatality_rate}%")
print(f"Not fatal rate: {round(100 - fatality_rate, 2)}%")
```

```python
# We are using 'values = year' to tell pandas which rows to count since they all have a year entry
pivot = pd.pivot_table(
    shark_clean_usa_filtered_df,
    index='type',
    columns='fatal y/n',
    values='year',
    aggfunc='count',
    fill_value=0
)

pivot['Total'] = pivot.sum(axis=1)

# Divide the fatal = y entries by the total amount (rounded to 2 decimal places)
pivot['Fatality Rate (%)'] = (pivot['y'] / pivot['Total'] * 100).round(2)

pivot.sort_values('Total')
```

```python
# We want to see if there is a correlation between certain species and fatality rates

pivot = pd.pivot_table(
    shark_clean_usa_filtered_df,
    index='species',
    columns='fatal y/n',
    values='year',
    aggfunc='count',
    fill_value=0
)

pivot['Total'] = pivot.sum(axis=1)
# Divide the fatal = y entries by the total amount (rounded to 2 decimal places)
pivot['Fatality Rate (%)'] = (pivot['y'] / pivot['Total'] * 100).round(2)

pivot.sort_values('Total')
```

```python
# Cleaning up Year column
# Step 1 - Dropping rows where year = 0.0 as they are not valid entries
shark_clean_usa_filtered_df = shark_clean_usa_filtered_df[shark_clean_usa_filtered_df['year'] != 0.0]

# Step 2 - Converting year from float to integer now that invalid values are removed
shark_clean_usa_filtered_df['year'] = shark_clean_usa_filtered_df['year'].astype(int)

# Step 3 - Verifying
print(shark_clean_usa_filtered_df['year'].dtype)
print(shark_clean_usa_filtered_df['year'].value_counts().sort_index())
```

```python
# Finding 1 - Total incidents per year
# Focusing on 1900+ where data coverage is reliable
incidents_per_year = shark_clean_usa_filtered_df[
    shark_clean_usa_filtered_df["year"] >= 1900
].groupby("year").size()

print(incidents_per_year.tail(27))
```

```python
# Finding 1 - Total incidents before 1975 vs total incidents after 1975
# Focusing on the last 100 years: 1925 - 2025

before_1975 = shark_clean_usa_filtered_df[
    (shark_clean_usa_filtered_df["year"] >= 1925) &
    (shark_clean_usa_filtered_df["year"] < 1975)
]

print(len(before_1975))

after_1975 = shark_clean_usa_filtered_df[
    (shark_clean_usa_filtered_df["year"] >= 1975) &
    (shark_clean_usa_filtered_df["year"] < 2025)
]

print(len(after_1975))
```

```python
# Finding 2 - Fatal vs Non-fatal per year
# If sharks were becoming more aggressive, fatality rate should rise over time
# A flat or declining fatality rate supports our narrative
fatal_trend = shark_clean_usa_filtered_df[
    shark_clean_usa_filtered_df['year'] >= 1900
].groupby(['year', 'fatal y/n']).size().reset_index(name= "count")
print(fatal_trend.tail(27))
```

```python
# Looking at the values and cleaning up the column Sex
shark_clean_usa_filtered_df["sex"].value_counts()
```

```python
#Cleaning column Sex
# Step 1 - Filling NaN values with 'Unknown' before stripping
shark_clean_usa_filtered_df['sex'] = shark_clean_usa_filtered_df['sex'].fillna('Unknown')

# Step 2 - Removing whitespace from Sex column
shark_clean_usa_filtered_df['sex'] = shark_clean_usa_filtered_df['sex'].str.strip()

# Step 3 - Keeping only M and F, everything else becomes Unknown
condition_sex = (shark_clean_usa_filtered_df['sex'] != 'M') & (shark_clean_usa_filtered_df['sex'] != 'F')
shark_clean_usa_filtered_df.loc[condition_sex, 'sex'] = 'Unknown'

# Displaying the cleaning result
shark_clean_usa_filtered_df['sex'].value_counts()
```

```python
# Finding 1 - Gender vs Fatality Rate
# Do fatal incidents more towards one gender?
gender_fatal_pivot = pd.pivot_table(
    shark_clean_usa_filtered_df[shark_clean_usa_filtered_df['sex'].isin(['M', 'F'])],
    index='sex',
    columns='fatal y/n',
    values='year',
    aggfunc='count',
    fill_value=0
)

gender_fatal_pivot['Total'] = gender_fatal_pivot.sum(axis=1)
gender_fatal_pivot['Fatality Rate (%)'] = (gender_fatal_pivot['y'] / gender_fatal_pivot['Total'] * 100).round(2)

print('Fatality rate by gender:')
print(gender_fatal_pivot)
```

```python
# Finding 2 - Gender vs Provoke
# Do fatal incidents more towards one gender?
gender_type_pivot = pd.pivot_table(
    shark_clean_usa_filtered_df[shark_clean_usa_filtered_df['sex'].isin(['M', 'F'])],
    index='sex',
    columns='type',
    values='year',
    aggfunc='count',
    fill_value=0
)

gender_type_pivot['Total'] = gender_type_pivot.sum(axis=1)
gender_type_pivot['Provoke Rate (%)'] = (gender_type_pivot['Provoked'] / gender_type_pivot['Total'] * 100).round(2)

print('Provoke rate by gender:')
print(gender_type_pivot)
```

```python
# Looking at the raw activity column

print("Top 20 raw values:")
print()
print(shark_clean_usa_filtered_df["activity"].value_counts().head(20))
```

```python
# Step 1 - Create a classification function to group activities
# We decided to define a function so the logic is reusable and readable
# The function checks each activity value and assigns it to a category

def group_activity(activity):
    # Cleaning null values first
    if pd.isnull(activity):
        return "Unknown"
    
    activity = activity.lower()
    
    if "surf" in activity:
        return "Surfing"                  
    elif "swim" in activity:
        return "Swimming"
    elif "spear" in activity or "fish" in activity:
        return "Fishing"
    elif "wade" in activity or "wading" in activity or "standing" in activity:
        return "Wading / Standing"
    elif "div" in activity or "snorkel" in activity or "scuba" in activity:
        return "Diving / Snorkeling"
    elif "kayak" in activity or "paddle" in activity or "canoe" in activity or "boat" in activity:
        return "Boating/Paddling"
    elif "boogie" in activity or "body board" in activity:
        return "Bodyboarding"
    else:
        return "Other"

# Step 2 - Applying the function to create a new column called activity_group
shark_clean_usa_filtered_df["activity_group"] = shark_clean_usa_filtered_df["activity"].apply(group_activity)

# Verifying the result
print("Activity group breakdown:")
print(shark_clean_usa_filtered_df["activity_group"].value_counts())
```

```python
# Step 3 - Displaying the "activity_group" incident count and percentage
# We are showing which activities are most common in shark incidents.

# Finding 1 - High numbers in Surfing and Swimming supports our narrative: "Humans entering shark territory, not sharks hunting humans."

activity_counts = shark_clean_usa_filtered_df["activity_group"].value_counts()
activity_pct = (shark_clean_usa_filtered_df["activity_group"].value_counts(normalize=True) * 100).round(1)

activity_summary = pd.DataFrame({
    "Incidents": activity_counts,
    "Percentage (%)": activity_pct
})

print("Activity breakdown:")
print(activity_summary)
```

```python
# Finding 2 - Activity vs Fatality Rate 
# The activitiy with highest incidents (Surfing) have the lowest Fatality rate with 1.80%.

activity_pivot = pd.pivot_table(
    shark_clean_usa_filtered_df,
    index="activity_group",
    columns="fatal y/n",
    values="year",
    aggfunc="count",
    fill_value=0
)

activity_pivot["Total"] = activity_pivot.sum(axis=1)
activity_pivot["Fatality Rate (%)"] = (activity_pivot["y"] / activity_pivot["Total"] * 100).round(2)

print("Activity vs Fatality Rate:")
print(activity_pivot.sort_values("Total", ascending=False))
```
